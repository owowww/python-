```python
#請印出1~100的質數，並顯示總共有幾個
total=0
for num in range(2,101):
    check = True
    for i in range(2,num):
        if(num%i==0):
            check = False
            #print("no")
            break
    if check == True:
        print("{}是質數".format(num))
        total=total+1

print("總共有{}個數".format(total))
```

```python
#檢查最低安全數量
import json

from urllib.request import urlopen
u = urlopen("http://data.ntpc.gov.tw/api/v1/rest/datastore/382000000A-000352-001")

resp = json.load(u)
keyword = input("請輸入要查詢的站名或關鍵字")
for i in range( len(resp["result"]["records"])):
    if( keyword in resp["result"]["records"][i]["sna"]): # in
        print(resp["result"]["records"][i]["sna"],end="")
        print(resp["result"]["records"][i]["sbi"])

for i in range( len(resp["result"]["records"])):
    rate = int(resp["result"]["records"][i]["sbi"]) / int(resp["result"]["records"][i]["tot"])
    if(rate <0.3):
        print("站名:{}，目前可借數量{}，車格數量{}，比例{}".format(resp["result"]["records"][i]["sna"],resp["result"]["records"][i]["sbi"],resp["result"]["records"][i]["tot"],rate))
```
